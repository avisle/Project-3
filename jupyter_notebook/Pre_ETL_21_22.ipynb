```python
# Import dependencies
import psycopg2
from sqlalchemy import create_engine, text
from config import username, password, hostname, port, db
```

```python
# Create engine
engine = create_engine(f'postgresql+psycopg2://{username}:{password}@{hostname}:{port}/{db}')
```

```python
# Add table as pandas dataframe (df)
table_df = pd.read_csv("Project_Resources/clean_exports/flu_data_all.csv") #Desired CSV file name
```

```python
table_df
```

```python
table_df = table_df.fillna(0)
table_df
```

```python
# Insert data into SQL database
table_df.to_sql('fluvaccall', con=engine, if_exists='replace')
```

```python
with engine.connect() as connection:
    result = connection.execute(text("SELECT * FROM fluvaccall"))
    for row in result:
        print(row)
```

```python
import plotly.express as px
```

```python
# Filter variables
season = '2018-19'
query = 'SELECT "90%_HCP_Vaccinated", count("90%_HCP_Vaccinated") FROM allfluvacc WHERE flu_season = ' + f"'{season}'" + 'GROUP by "90%_HCP_Vaccinated" ORDER BY count("90%_HCP_Vaccinated")'
sql_query = text(query)
conn = engine.connect()
pie_df = pd.read_sql(sql_query, conn) #TABLE NAME
pie_df
```

```python
season = '2020-21'
query = 'SELECT "90%_HCP_Vaccinated", count("90%_HCP_Vaccinated") FROM allfluvacc WHERE flu_season = ' + f"'{season}'" + 'GROUP by "90%_HCP_Vaccinated" ORDER BY count("90%_HCP_Vaccinated")'
sql_query = text(query)
conn = engine.connect()
pie2_df = pd.read_sql(sql_query, conn) #TABLE NAME
pie2_df
```
