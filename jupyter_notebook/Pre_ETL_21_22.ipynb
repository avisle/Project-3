````python
# Import dependencies
import pandas as pd
````

````python
# Read the data into a Pandas DataFrame
flu_data_21_22_df = pd.read_csv('Resources/raw_csvs/flu_data_21_22.csv')
flu_data_21_22_df
````

````python
# List columns
flu_data_21_22_df.columns
````

````python
# Renaming columns for SQL
new_columns = {
    '_id': 'ID',
    'Influenza_Season': 'Flu_Season',
    'CDPH_HAI_HCP_FLU_VACCINATION': 'Program',
    'order_this_season': 'Mask_Worn',
    'HCP_Total_Number_Working': 'Total_HCP_Vaccinated',
    'AllHCP_FluVaccPerc_r': 'Percentage_HCP_Vaccinated',
    'Emp_Number_Vaccinated': 'Total_Hospital_Employees_Vaccinated',
    'Emp_Number_Working': 'Total_Hospital_Employees_Working',
    'Emp_FluVaccPerc_r': 'Percentage_Hospital_Employees_Vaccinated',
    'Cont_Number_Vaccinated': 'Total_Number_Contractors_Vaccinated',
    'Cont_Number_Working': 'Total_Number_Contractors_Working',
    'Con_FluVaccPerc_r': 'Percentage_Contractors_Vaccinated',
    'Lip_Number_Vaccinated': 'Total_Number_LIP_Vaccinated',
    'Lip_Number_Working':'Total_Number_LIP_Working',
    'LIP_FluVaccPerc_r': 'Percentage_LIP_Vaccinated',
    'StudVol_Number_Vaccinated': 'Total_Number_Volunteers_Vaccinated',
    'StudVol_Number_Working': 'Total_Number_Volunteers_Working',
    'StudVolFluVaccPerc_r': 'Percentage_Volunteers_Vaccinated',
    'Met_2020Goal': '90%_HCP_Vaccinated',
    'Notes':'Hospital_Reported_Data'    
}

flu_data_21_22_df = flu_data_21_22_df.rename(columns=new_columns)

flu_data_21_22_df.columns
````

````python
# Dropping columns
flu_data_21_22_df_cleaned = flu_data_21_22_df.drop(['Program','Mask_Worn','Hospital_Reported_Data'], axis=1)

flu_data_21_22_df_cleaned
````

````python
# Export the DataFrame as a CSV file. 
flu_data_21_22_df_cleaned.to_csv("./Project_Resources/clean_exports/flu_data_clean_21_22.csv", index=False)
````

````python
# Import dependencies

import psycopg2
from sqlalchemy import create_engine, text
from config import username, password, hostname, port, db
````

````python
# Create engine
engine = create_engine(f'postgresql+psycopg2://{username}:{password}@{hostname}:{port}/{db}')
````

````python
# Add table as pandas dataframe (df)
table_df = pd.read_csv("Project_Resources/clean_exports/flu_data_clean_21_22.csv")
````

````python
# Insert data into SQL database
table_df.to_sql('FluVacc2122', con=engine, if_exists='replace')
````
