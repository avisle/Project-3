```python
#Import necessary dependencies/libraries
import pandas as pd
pd.set_option('max_colwidth', 400)
```

```python
#Convert the raw data from the csv specifying the directory path into a Pandas Dataframe
flu_data_18_19_df = pd.read_csv('Resources/raw_csvs/flu_data_18_19.csv')
flu_data_18_19_df
```

```python
#Show the column names of the dataframe
flu_data_18_19_df.columns
```

```python
#Change the column names to be better more understandable
new_columns = {
    '_id': 'ID',
    'Influenza_Season': 'Flu_Season',
    'CDPH_HAI_HCP_FLU_VACCINATION': 'Program',
    'Mask Required': 'Mask_Worn',
    'HCP_Total_Number_Vaccinated': 'Total_Healthcare_Professionals_Vaccinated',
    'HCP_Total_Number_Working': 'Total_Healthcare_Professionals_Working',
    'HCP_Percent_Vaccinated': 'Percentage_Healthcare_Professionals_Vaccinated',
    'Emp_Number_Vaccinated': 'Total_Hospital_Employees_Vaccinated',
    'Emp_Number_Working': 'Total_Hospital_Employees_Working',
    'Emp_Percent_Vaccinated': 'Percentage_Hospital_Employees_Vaccinated',
    'Cont_Number_Vaccinated': 'Total_Number_Contractors_Vaccinated',
    'Cont_Number_Working': 'Total_Number_Contractors_Working',
    'Cont_Percent_Vaccinated': 'Percentage_Contractors_Vaccinated',
    'Lip_Number_Vaccinated': 'Total_Number_LIP_Vaccinated',
    'Lip_Number_Working': 'Total_Number_LIP_Working',
    'Lip_Percent_Vaccinated': 'Percentage_LIP_Vaccinated',
    'StudVol_Number_Vaccinated': 'Total_Number_Volunteers_Vaccinated',
    'StudVol_Number_Working': 'Total_Number_Volunteers_Working',
    'StudVol_Percent_Vaccinated': 'Percentage_Volunteers_Vaccinated',
    'On Track to Achieve 90% by 2020 (>=87% in 2018-19)': '90%_HCP_Vaccinated',
    'Notes': "Hospital_Reported_Data"
}

#Rename the columns in the dataframe with new names in new_columns
flu_data_18_19_df = flu_data_18_19_df.rename(columns=new_columns)

#Show the new columns
flu_data_18_19_df.columns
```

```python
#Remove the columns that aren't needed and output the dataframe
flu_data_18_19_cleaned_df = flu_data_18_19_df.drop(['Program','Mask_Worn','Hospital_Reported_Data'], axis=1)
flu_data_18_19_cleaned_df
```

```python
#Export the cleaned data
flu_data_18_19_cleaned_df.to_csv("Resources/clean_exports/flu_data_cleaned_18_19.csv", index=False)
```
